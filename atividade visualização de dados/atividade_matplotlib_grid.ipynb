```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando variavel x
X = np.linspace(0,5,10)
```

```python
#criando funções y(x) e z(x)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#criando grid
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid( linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

```python
#alterando cor do fundo
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
ax = plt.axes()
ax.set_facecolor('purple')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
