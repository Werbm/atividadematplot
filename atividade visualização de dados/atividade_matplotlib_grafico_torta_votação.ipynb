```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
votos = np.array([842.201, 488.775, 553.424, 424.307, 272.500, 381.512, 261.386])
candidatos = ['Candidato A', 'Candidato B', 'Candidato C', 'Candidato D', 'Candidato E', 'Candidato F', 'Candidato G']
cores=['yellow', 'purple', 'blue', 'magenta', 'green','lightblue', 'brown']
# o atributo explode indica que fatia do gráfico será destacada. 
explode = (0.1, 0, 0, 0, 0, 0, 0)  # explode a primeira fatia do gráfico
```

```python
#atribuindo um título ao gráfico
plt.title('Eleição 2020')
plt.pie(votos, explode=explode, labels=candi, colors=cores, autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(candi, bbox_to_anchor=(1.3, 1.3),loc='upper right')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
