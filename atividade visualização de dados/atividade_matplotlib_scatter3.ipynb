```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando variável X
X = np.linspace(0,5,10)
```

```python
#Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
#alterando tamanho do ponto
plt.scatter(X,Y,s=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
plt.scatter(X,Y,s=10,label='Y(X)')
plt.scatter(X,Z,s=40,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,s=10,label='Y(X)', alpha=0.8)
plt.scatter(X,Z,s=40,label='Z(X', alpha=0.4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
