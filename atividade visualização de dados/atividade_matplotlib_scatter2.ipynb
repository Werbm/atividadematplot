```python
#importando bibliotecas
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#criando variavel x
X = np.linspace(0,5,10)
```

```python
#criando função Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#gráfico de y em função de x
plt.scatter(X,Y)
plt.show()
```

```python
#alterando simbolo
plt.scatter(X,Y,marker='.')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
#dois gráficos com simbolos diferentes
plt.scatter(X,Y,marker='p')
plt.scatter(X,Z,marker='v')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
