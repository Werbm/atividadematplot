```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels =["A1", "A2", "A3", "A4", "A5"]
```

```python
valores = [10,20,32,50,15]
```

```python
plt.bar(labels,valores)
```

```python
plt.bar(labels,valores)
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
plt.bar(labels,valores,color='purple')
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
plt.barh(labels,valores,color='purple')
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
plt.bar(labels,valores,color='purple')
plt.xlabel('ID')
plt.ylabel('Valor')

for x,y in zip(labels,valores):
    plt.text(x, y+0.05, '%d' % y, ha='center', va='bottom')
```

```python
valores2 = [2,10,20,30,40]
```

```python
largura = 0.3
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, color='purple', width=largura, label='Barra 1')
plt.bar(pos2, valores2, color='blue', width=largura, label='Barra 2')
plt.legend()
```

```python
plt.bar(pos1,valores, label = "Barra 1")
plt.bar(pos1,valores2,label = "Barra 2", bottom = valores)
```

```python
plt.barh(pos1,valores, label = "Barra 1")
plt.barh(pos1,valores2, label = "Barra 2", left = valores)
```
