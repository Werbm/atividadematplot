```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando variavel x
X = np.linspace(0,5,10)
```

```python
#criando funções y(x) e z(x)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#grafico de y e função x
plt.scatter(X,Y)
plt.show()
```

```python
#inserindo nomes nos eixos
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
#inserindo titulo no gráfico
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
#dois gráficos na mesma imagem
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
#alterando cores
plt.scatter(X, Y,color= 'black')
plt.scatter(X,Z,color= 'red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
plt.scatter(X,Y,color='black',label='Y(X)')
plt.scatter(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
