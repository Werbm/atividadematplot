```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando variavel x com numeros aleatorios
X = np.random.rand(2000)
```

```python
#criando histograma
plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

```python
#alterando numero de bins
plt.hist(X, bins=10)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=10')
plt.show()
```

```python
plt.hist(X,bins=20)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=20')
plt.show()
```

```python
plt.hist(X,bins=30)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=30')
plt.show()
```

```python
plt.hist(X,bins=50,color='purple')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X,bins=50,color='purple', edgecolor='blue')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
#multiplos histogramas na mesma imagem
x1 = np.random.normal(0,1,size = 800)
x2 = np.random.normal(3,1,size =800)
x3 = np.random.normal(5,1,size = 800)
```

```python
plt.hist(x1,bins=10,color='purple')
plt.hist(x2,bins=10,color='blue')
plt.hist(x3,bins=10,color='pink')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos Histogramas')
plt.show()
```

```python
#distribui√ß√£o cumulativa
X = np.random.rand(2000)
```

```python
plt.hist(X,bins=20,color='purple', edgecolor='blue', cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma Cumulativo')
plt.show()
```
