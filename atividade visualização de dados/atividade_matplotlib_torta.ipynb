```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando labels
labels = ["A1", "A2", "A3", "A4", "A5"]
```

```python
#criando lista com valores
valores = [10,45,32,56,24]
```

```python
#criando gr√°fico de barras
plt.pie(valores,labels=labels)
plt.show()
```

```python
#inserindo porcentagem
plt.pie(valores,labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#explode
explode = [0,0.1,0,0,0]
```

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%', explode=explode)
plt.show()
```
