```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando dados
x = [10, 14, 25, 30, 40]
y = [10, 23, 35, 32, 40]
```

```python
#criando uma figura e um eixo
fig, ax = plt.subplots()
#plotar os dados com as personalizações
ax.plot(x, y, color='grey', linestyle='--', marker='o', label='dados')
#adicionando rótulos
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Y(X)')
ax.legend(loc='upper left')
plt.show()
```
