```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando variável x
X= np.linspace(0,5,10)
```

```python
#criando funções y(x) e z(x)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#gráfico de Y em função de X
plt.plot(X,Y)
plt.show()
```

```python
#inserindo nomes nos eixos
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
#dois gráficos na mesma imagem
plt.plot(X,Y)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
#alterando cores
plt.plot(X,Y,color='black')
plt.plot(X,Z,color='red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
plt.plot(X,Y, color='black',label='Y(X)')
plt.plot(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
