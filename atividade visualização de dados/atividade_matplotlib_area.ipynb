```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10,5,101)
```

```python
Y = (X-2.0)*(X+2.0)
```

```python
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(X, Y)
plt.show()
```
